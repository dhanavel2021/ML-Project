```python
import pandas as pd
```

```python
data=pd.read_csv("/content/HousePrices.csv")
```

```python
data=data.drop(["Id"],axis=1)
```

```python
dc = data.columns[data.dtypes=='object']
```

```python
dc
```

```python
dn = data.columns[data.dtypes!='object']
```

```python
dn
```

```python
dfc=data[dc]
```

```python
dfn=data[dn]
```

```python
dfc.isna().sum()
```

```python
dfn.isna().sum()
```

```python
dfc = dfc.apply(lambda x: x.fillna(x.value_counts().index[0])) 
```

```python
dfc
```

```python
dfn = dfn.apply(lambda x: x.fillna(x.median()))
```

```python
dfn
```

```python
dfcc=pd.get_dummies(dfc)
dfcc
```

```python
df=pd.concat([dfcc,dfn],join='outer',
                      axis=1)
```

```python
df
```

```python
x=df.drop(["Property_Sale_Price"],axis=1)
y=df["Property_Sale_Price"]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
m=LinearRegression()
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
xtrain,xtest,ytrain,ytest=tts(x,y,test_size=.30,random_state=0)
```

```python
m.fit(xtrain,ytrain)
```

```python
p=m.predict(xtrain)
```

```python
r=m.predict(xtest)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
r2_score(ytrain,p)
```

```python
r2_score(ytest,r)
```

```python
mean_squared_error(ytrain,p)
```

```python
mean_squared_error(ytest,r)
```
